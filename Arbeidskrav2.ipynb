```python
# Oppgave 5
#Bibliotek
import numpy as np

#Kode

def beregn_areal_og_omkrets(a, b):
    areal_trekant = 0.5 * a * b
    radius = a / 2
    areal_halvsirkel = 0.5 * np.pi * radius ** 2
    total_areal = areal_trekant + areal_halvsirkel

    hypotenus = np.sqrt(a**2 + b**2)
    bue_halvsirkel = np.pi * radius
    ytre_omkrets = b + hypotenus + bue_halvsirkel

    return total_areal, ytre_omkrets


a = float(input("Skriv inn lengden a (diameteren til halvsirkelen): "))
b = float(input("Skriv inn lengden b (katet i trekanten): "))


areal, omkrets = beregn_areal_og_omkrets(a, b)


print("Resultater for figur med a =", a, "og b =", b)
print("Areal:", round(areal, 2), "kvadrat-enheter")
print("Ytre omkrets:", round(omkrets, 2), "enheter")
```

```python
# Oppgave 6
# Bibliotek
import numpy as np
import matplotlib.pyplot as plt

# Kode
x = np.linspace(-10, 10, 200)


y = -x**2 - 5


plt.figure(figsize=(8, 5))
plt.plot(x, y, label='f(x) = -x² - 5', color='blue')
plt.title('Plot av f(x) = -x² - 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
